```python
# Importing the Libraries

import yfinance as yf

import pandas as pd
import numpy as np
```

```python
# Starting from Microsoft
msft = yf.Ticker("MSFT")

# get stock info
# print(msft.info)
```

```python
# get historical market data
hist = msft.history(period="40d")
```

```python
hist.head()
```

```python
len(hist.index)
```

```python
hist.tail()
```

```python
df = hist.copy()
```

```python
df.head()
```

```python
df.sort_values(['Date'],ascending=False,inplace=True)
```

```python
df.head()
```

```python
df['Rel Change(%) - 1 day']=round((df['Open']-df['Open'].shift(-1))/df['Open']*100,4)
```

```python
df['Rel Change(%) - 7 day(1 week)']=round((df['Open']-df['Open'].shift(-7))/df['Open']*100,4)
```

```python
df['Rel Change(%) - 30 day(1 month)']=round((df['Open']-df['Open'].shift(-30))/df['Open']*100,4)
```

```python
df.head(10)
```

```python
# Saving as CSV
df.to_csv("msft.csv")
```
